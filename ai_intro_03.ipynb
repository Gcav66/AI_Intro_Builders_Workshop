```python
import numpy as np
import matplotlib.pyplot as plt

# Let's create our data using NumPy arrays.
# X is our feature (video length in minutes).
X = np.array([5, 8, 9, 11, 14])

# y is our label (views in thousands).
y = np.array([11, 15, 19, 21, 29])

# Now, let's create a scatter plot to visualize the data.
# plt.figure() creates a canvas for our plot.
plt.figure(figsize=(8, 6))

# plt.scatter() plots our data points.
plt.scatter(X, y, color='blue', label='Actual Data')

# Add some labels to make our plot understandable.
plt.title("CodeCrafted's YouTube Views vs. Video Length")
plt.xlabel("Video Length (minutes)")
plt.ylabel("Views (in thousands)")
plt.legend()
plt.grid(True)

# plt.show() displays the plot.
plt.show()
```

```python
# First, we need to reshape our X data into a 2D array.
# The .reshape(-1, 1) method does this for us.
X_reshaped = X.reshape(-1, 1)
print("Original X shape:", X.shape)
print("Reshaped X shape:", X_reshaped.shape)


# --- The Three Magic Lines of Scikit-learn ---

# 1. Import the model we want to use.
from sklearn.linear_model import LinearRegression

# 2. Instantiate the model. This creates our "learner" object.
model = LinearRegression()

# 3. Fit the model to our data. This is the training step.
# The model learns the relationship between X_reshaped and y.
model.fit(X_reshaped, y)

print("\nTraining complete! The model has learned the line of best fit.")
```

```python
# Get the learned parameters (m and b)
slope_m = model.coef_[0]
intercept_b = model.intercept_

print(f"\nThe model learned the following equation:")
print(f"Views = {slope_m:.2f} * VideoLength + {intercept_b:.2f}")
```

```python
# (Assuming you have run the code from the previous examples)

# --- Making a Prediction ---
# Let's predict the views for a new 12-minute video.
# We need to put our new data point into a 2D array as well.
new_video_length = np.array([[12]])

# Use the .predict() method to get the model's prediction.
predicted_views = model.predict(new_video_length)

print(f"\nPrediction for a {new_video_length[0][0]}-minute video: {predicted_views[0]:.2f} thousand views.")


# --- Visualizing the Model ---
# We can create a set of x-values to draw our line.
x_line = np.linspace(4, 15, 100).reshape(-1, 1)

# For each of those x-values, we predict the corresponding y-value.
y_line = model.predict(x_line)

# Plot the original data again.
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Actual Data')

# Plot the regression line our model learned.
plt.plot(x_line, y_line, color='red', linewidth=2, label='Line of Best Fit')

plt.title("Linear Regression Model for YouTube Views")
plt.xlabel("Video Length (minutes)")
plt.ylabel("Views (in thousands)")
plt.legend()
plt.grid(True)
plt.show()
```
